### 선그래프

```r
# 3종류 데이터의 선그래프
abc <- c(260,300,250,280,310)
def <- c(180,200,210,190,170)
ghi <- c(210,250,260,210,270)
```

```r
par(family = "AppleGothic")
par(bg = 'white')
plot(
    abc,
    type = "o",
    col = "red",
    ylim = c(0,400), #y축 범위조절
    axes = F, #x,y의 좌표 프레임 제거
    ann = F #x,y의 Label 프레임 제거
    
)



# 'x'축과 'x'축의 라벨을 설정
axis(1, at=1:5, lab=c("서울", "대전", "대구", "광구", "원주"))
# 'y'축과 'y'축의 라벨을 설정
axis(2, ylim = c(0,400))



# main title 지정
title(main='과일 판매량', col.main='red', cex.main=2)
# x축 title 지정
title(xlab="지역", col.lab = "blue", cex.lab=2)
# y축 title 지정
title(ylab = "판매량", col.lab= "blue")



# 그래프 추가
lines(
    def,
    type = "o",
    col = "green",
    lty = 2    
)

lines(
    ghi,
    type = "o",
    col = "blue",
    lty = 2,
    pch = 22 
)

# 범례
legend(
    4,   #(x축 4번째인 광주위치)
    400, #(y축의 400번 위치)
    c("야구장","축구장","해변가"),
    # fill = c("red","green","blue")  (네모칸 색으로 표시)
    col = c("red","green","blue"),
    pch = c(21,21,22),
    lty = c(1,2,2),
    cex = 0.8

)
```

```r
### 파이차트
par(bg="white")
sales <- c(210,110,400,550,600,130)
pie(sales)
```

```r
par(bg="white")
sales <- c(210,110,400,550,600,130)
week <- c("월","화","수","목","금","토")
ratio <- sales / sum(sales) * 100
label <- paste(week, "\n", ratio, "%")
label

pie(
    sales,
    main = "영업 실적 파이 차트",
    col = rainbow(length(sales)),
    labels = label)
```

```r
par(bg="white")
sales <- c(210,110,400,550,600,130)
week <- c("월","화","수","목","금","토")
ratio <- sales / sum(sales) * 100
label <- paste(week, "\n", ratio, "%")
label

pie3D(
    sales,
    main = "영업 실적 파이 차트",
    col = rainbow(length(sales)),
    labels = label,
    explode = 0.05,  #모양 쪼개기
    labelcex = 0.8
)
```
