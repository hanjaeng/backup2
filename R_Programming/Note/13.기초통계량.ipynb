### 수학 함수의 시각화

```r
: y = x^2
```

```r
f <- function(x){
    return(x^2)
}
```

```r
f(3)
```

```r
x <- 1:5
f(x)
```

```r
x <- seq(-5,5,0.1) #-5에서 5까지 0.1씩
x
```

```r
par(bg = "white")
plot(
    x,
    f(x),
    type='l'
)
```

#

```r
# y = x^2 + 2x +3 시각화

k <- function(x){
    return(x^2 + 2*x + 3)
}

x <- seq(-5, 5, 0.1) #-5에서 5까지 0.1씩

par(bg = "white")

plot(
    k,
    f(x),
    type='l'
)
```

```r
# y = e^(-x) - 4x^2 + x + 3|x| -3

g <- function(x){
    return(e^(-x) - 4*x^2 + x + 3*x -3)
}

x <- seq(-5, 5, 0.1) #-5에서 5까지 0.1씩

par(bg = "white")

plot(
    g,
    f(x),
    type='l'
)
```

### 기초 통계량

```r
cafe <- read.csv("../Data/coffee.csv")
head(cafe)
```

```r
# 커피의 빈도수
table(cafe$Coffee)
```

```r
# 범위 빈도수
coffee_range <- seq(0,60,10) #범위
coffee_range
cut(cafe$Coffee, breaks = coffee_range)  #breaks 범위 
table(cut(cafe$Coffee, breaks = coffee_range))  #'('초과,  ']'이하
table(cut(cafe$Coffee, breaks = coffee_range, right=F))   #'['이상,  ')'미만
```

```r
# 줄기 - 잎 그림
stem(cafe$Coffee)
```

```r
# 커피와 쥬스판매량의 변동계수를 구하여 메뉴를 선택하자(변동계수가 적어야 변화가 적음)
table(cafe$Coffee)
table(cafe$Fruit.Cup)

stem(cafe$Coffee)
stem(cafe$Fruit.Cup)
```

```r
sd(cafe$Coffee)  #분산
sd(cafe$Fruit.Cup)

sqrt(var(cafe$Coffee)) #표준편차
sqrt(var(cafe$Fruit.Cup))
```
